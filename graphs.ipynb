```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%load_ext autoreload
%autoreload 2
```

```python
fig, ax = plt.subplots()

for run in [
    "transfer_False__data_ignored_0.csv",
    "transfer_False__data_ignored_40.0.csv",
    "transfer_False__data_ignored_80.0.csv",
]:
    stats = pd.read_csv(f".logs/{run}", header=None)

    train = stats.iloc[:, 1]
    val = stats.iloc[:, 2]

    ax.plot(train, label=f"Train: {run.split('_')[-1][:-4]}% of data discarded")
    ax.plot(val, label=f"Val:    {run.split('_')[-1][:-4]}% of data discarded")

ax.set_title("Without pretraining on the vegetable set")
ax.set_xlabel("Epochs")
ax.set_ylabel("Cross Entropy Loss")
ax.set_ylim(0, 1.5)
ax.set_xlim(0, 40)

ax.legend(loc="upper right")
plt.show()
```
